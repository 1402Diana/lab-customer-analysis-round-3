```python
#Show a plot of the response rate by the sales channel.
sns.countplot(x="sales channel", hue="response", data=df)

# Add the title et the labels.
plt.title('Count of Responses by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Count')

# Show the grath.
plt.show()
```
